```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_csv('fara.csv')
```

```python
df.head()
```

```python
node_list = list()
edge_list = list()
for country in df['Country/LocationRepresented'].unique():
  node_list.append({
      'id' : country,
      'Label' : country,
      'type' : 'country',
      'country' : country
  })
  for index, record in df[df['Country/LocationRepresented'] == country].iterrows():
    # record = dict(record)
    node_list.append({
      'id' : record['Foreign Principal'],
      'Label' : record['Foreign Principal'],
      'type' : 'foreign_principal',
      'country' : country
  })
    

    node_list.append({
      'id' : record['Registrant'],
      'Label' : record['Registrant'],
      'type' : 'registrant',
      'country' : country
  })
    edge_list.append({
        'source' : country,
        'target' : record['Foreign Principal']
    })
    edge_list.append({
        'source' : record['Foreign Principal'],
        'target' : record['Registrant']
    })
     



ndf = pd.DataFrame(node_list)
edf = pd.DataFrame(edge_list)
```

```python
edf.to_csv('edge_list_fara.csv', index = False)
ndf.to_csv('node_list_fara.csv', index = False)
```

```python
bi_node_list = list()
bi_edge_list = list()
for country in df['Country/LocationRepresented'].unique():
  bi_node_list.append({
      'id' : country,
      'Label' : country,
      'type' : 'country',
      'country' : country
  })
  for index, record in df[df['Country/LocationRepresented'] == country].iterrows():
    bi_node_list.append({
      'id' : record['Registrant'],
      'Label' : record['Registrant'],
      'type' : 'registrant',
      'country' : country
  })
    bi_edge_list.append({
        'source' : country,
        'target' : record['Registrant']
    })
     



bi_ndf = pd.DataFrame(bi_node_list)
bi_edf = pd.DataFrame(bi_edge_list)
```

```python
bi_ndf.to_csv('bi_node_list_fara.csv', index = False)
bi_edf.to_csv('bi_edge_list_fara.csv', index = False)
```
